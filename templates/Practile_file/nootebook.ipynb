```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("D:\Project-1\Sales-data.csv")
```

```python
df.head()
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
```

```python
df.info()
```

```python
df["Day"] = df["Order Date"].dt.day
df["Month"] = df["Order Date"].dt.month
df["Year"] = df["Order Date"].dt.year
df["Hours"] = df["Order Date"].dt.hour
df["Minute"] = df["Order Date"].dt.minute
```

```python
df.head()
```

```python
df2 = df.drop(columns=["Order Date","Unnamed: 0","Hour"])
```

```python
df2.rename(columns={"Order ID":"Order_ID","Quantity Ordered":"Quantity_Ordered","Price Each":"Price_Each"},inplace=True)
```

```python
df2.head()
```

```python
x = df2.drop(columns=["Sales","Product","Purchase Address","City"])
y = df2[["Sales"]]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_train_df = scaler.fit_transform(x_train)
x_test_df= scaler.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
reg.fit(x_train_df,y_train)
```

```python
y_pred = reg.predict(x_test_df)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
```

```python
print(r2_score(y_pred,y_test))
```

```python
import pickle
pickle.dump(reg,open("reg.pkl","wb"))
pickle.dump(scaler,open("scaler.pkl","wb"))
```
